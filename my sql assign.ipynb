database is an organised collection of data or information stored and accessed electronically.
small databases can be stored on a file system, while large databases are hosted on computer clusters or cloud storage.
a database is usuallly controlled by a database management system (DBMS).

diff btw sql and nosql databases:->

sql:-

      1) sql database is a table based databse.
      
      2) it is vertically scalable i.e, it scales up with a larger servers.
      
      3) it is primarilay called relational database management system (RDBMS).
      
      4) best suited for complex queries.
      
      5) it consist of structured data.
      
      6) examples:- mysql, mssql, oracle, etc.
      
nosql:-
        
        1) nosql databases can be document based, key-value pairs and graph databases.
        
        2) it is horizontally scalable i.e, it scales out accross commodity servers.
        
        3) it is primarily called as non relational or disturbed database management system (RDMS/DDMS).
        
        4) not so good for solving complex queries.
        
        5) it consist of unstructured data.
        
        6) examples:- mongodb, hbase, neo4j, cassandra, etc.

* Primary key:-> let's say we have a table named student which stores information of students. having columns rollno, name, phone, address.
it is possible that two or more students can have the same names. But each student must have a unique roll no. To fetch the phone no of a particular student who has roll no 2, we will go to the database and query it to return the information of roll no 2.

we can see that every person has a roll no. that is unique and we can directly identify a person based on his roll no. So primary key is a unique value used to identify a row in a table. The primary key can be anything in the table which is unique for a row. Here we can also take the phone no as the primary key because it is unique for every person.

syntax:-> CREATE TABLE Student(roll no INT, name VARCHAR(50),Phone VARCHAR(20), address VARCHAR(300),PRIMARY KEY (roll no));

points to be remembered while creating a primary key:- a) a table can have only one primary key.
                                                       b) the primary key is also denoted as Pk.
                                                       c) primary key cannot be null.
                                                       d) we can create the primary key by combining two or more columns.The resultant key is known as                                                           a composite key. Example: Phone_no + name can also act as primary key because it can                                                                   uniquely identify rows.
                                                       
                                                       
* Foreign key:-> it is a column or group of column that connects or make a relationship between two tables. let's understand about this key with an
example. suppose we want to store the performance of students about whom we were talking previously in primary key.
firstly we create a performance table which stores the performance of students as performanceid(PID), subject name, marks, total marks, roll no.

The performance table has a column named roll no which is the same as the primary key of the student table. This column will act as a foreign key for the performance table. Now we can find the performance of any student by querying the performance table and specifying the student roll no in the query.

syntax:-> CREATE TABLE Performance (PID INT, subject name VARCHAR(300), marks INT, PRIMARY KEY (PID), FOREIGN KEY (roll no) REFERENCES Student(rollno) );

points to be remembered while creating a foreign key:- a) it is also denoted as Fk.
                                                       b) it can have multiple null values.
                                                       c) table with a foreign key is called a child table/referencing table and the table with the                                                             primary key is called a parent table/referenced table.
